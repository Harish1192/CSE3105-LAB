```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Mall_Customers.csv")
data.tail(10)
```

```python
X=data.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5,init='k-means++',random_state=0)
y=km.fit_predict(X)
```

```python
y
```

```python
plt.scatter(X[y==0,0],X[y==0,1],s=100,c='blue')
plt.scatter(X[y==1,0],X[y==1,1],s=100,c='orange')
plt.scatter(X[y==2,0],X[y==2,1],s=100,c='yellow')
plt.scatter(X[y==3,0],X[y==3,1],s=100,c='red')
plt.scatter(X[y==4,0],X[y==4,1],s=100,c='purple')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='black')#Centroid
plt.xlabel("Annual Income")
plt.ylabel("Spending Money")
```
